```python
import datetime

import IPython.display
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas_datareader
import sklearn
import sklearn.linear_model
import sklearn.model_selection


# IPyhon.display で画像の読み込みができる
# IPython.display.Image('データ名', width=サイズ)
```

```python
# データウェアハウス
df_appl = pandas_datareader.data.DataReader('AAPL', 'yahoo', '2020-10-01')
df_fb = pandas_datareader.data.DataReader('FB', 'yahoo', '2020-10-01')
df_gold = pandas_datareader.data.DataReader('GLD', 'yahoo', '2020-10-01')
```

```python
df_appl.tail(2)
```

```python
# 統計学 Simple Moving Averege
# 過去14日間の平均
df_appl['SMA'] = df_appl['Close'].rolling(window=14).mean()
df_appl['Close'].plot(figsize=(15,6), color='red')
df_appl['SMA'].plot(figsize=(15,6), color='green')
plt.show()
```

```python
# データマイニング
df_appl['change'] = (((df_appl['Close'] - df_appl['Open'])) / (df_appl['Open']) * 100)
df_fb['change'] = (((df_fb['Close'] - df_fb['Open'])) / (df_fb['Open']) * 100)
df_gold['change'] = (((df_gold['Close'] - df_gold['Open'])) / (df_gold['Open']) * 100)
df_appl.tail(2).round(2)
```

```python
df_appl['Close'].plot(figsize=(15,6), color='red')
df_fb['Close'].plot(figsize=(15,6), color='blue')
plt.show()
```

```python
df_appl['Close'].plot(figsize=(15,6), color='red')
df_gold['Close'].plot(figsize=(15,6), color='gold')
plt.show()
```

```python
df_appl['change'].tail(100).plot(grid=True, figsize=(15,6), color='red')
df_fb['change'].tail(100).plot(grid=True, figsize=(15,6), color='blue')
df_gold['change'].tail(100).plot(grid=True, figsize=(15,6), color='gold')
plt.show()
```

```python
df_appl['label'] = df_appl['Close'].shift(-30)
df_appl.tail(40)
```

```python
# 機械学習

# ラベル行を削除したデータ
X = np.array(df_appl.drop(['label', 'SMA'], axis=1))
# 株価の乱高下が激しく扱うには悪いデータ
# 取りうる値の大小が著しく異なる特徴量を入れると結果が悪くなり、平均を引いて、標準偏差で割ってスケーリングする
X = sklearn.preprocessing.scale(X)

# 予測に使う過去30日のデータ
predict_data = X[-30:]
# 過去30日を取り除いた入力データ
X = X[:-30]
y = np.array(df_appl['label'])
# 過去30日を取り除いた正解ラベル
y = y[:-30]

# 訓練データとテストデータに分ける
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2)

lr = sklearn.linear_model.LinearRegression()
lr.fit(X_train, y_train)

accuracy = lr.score(X_test, y_test)
accuracy
```

```python
predict_data = lr.predict(predict_data)
predict_data
```

```python
df_appl['Predict'] = np.nan

last_data = df_appl.iloc[-1].name

one_day = 86400
next_unix = last_data.timestamp() + one_day

for data in predict_data:
    next_date = datetime.datetime.fromtimestamp(next_unix)
    next_unix += one_day
    df_appl.loc[next_date] = np.append([np.nan] * (len(df_appl.columns) - 1), data)

df_appl['Close'].plot(figsize=(15,6), color='green')
df_appl['Predict'].plot(figsize=(15,6), color='orange')
plt.show()
```
