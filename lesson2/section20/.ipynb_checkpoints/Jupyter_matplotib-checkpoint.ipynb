```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
# 引数に入れたリストは x座標 y座標 の順
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
# x が 0~6 まで y が 0~10 まで
plt.axis([0, 6, 0, 10])
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# 座標や大きさ、色、透明度などを指定する
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
area = np.random.rand(50) * 100
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
```

```python
objects = ('a', 'b', 'c', 'd', 'e', 'f')
y_pos = np.arange(len(objects))
value = [1, 2, 3, 4, 5, 6]

plt.bar(y_pos, value, alpha=0.5)
# xticks は x軸のメモリ
# 1, 2, 3... がそれぞれ 'a', 'b', 'c'...に対応する
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Title')
plt.show()
```

```python
labels = ('Python', 'C++', 'Ruby', 'Java')
sizes = [10, 20, 30, 40]
colors = ['red', 'green', 'yellow', 'blue']

# autopct の部分で浮動小数点の % を表示する
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
# plt.axis('equal') で形を整える
plt.axis('equal')
plt.show()
```

```python
y = [2, 4, 6, 8, 10]
y2 = [10, 11, 12, 13, 14]
x = np.arange(5)

plt.plot(x, y, 'rx', label='Y')
plt.plot(x, y2, label='Y2')
# legend で説明を付け加えることができる
plt.legend()
plt.show()
```
